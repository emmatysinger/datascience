# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {v^2} g\sin 2\theta$.

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta): 
    d = v**2/9.81*sin(radians(2*theta))
    return (d)
```

```python
rangeformula(100,45)
```

```python
angledegrees = range(91)
```

```python
rangefromdegrees = [rangeformula(100,x) for x in angledegrees]
```

```python
plt.plot(range(91), rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (º)')
plt.ylabel('Range (m)')
plt.axis([0,100,0,1200])
plt.show()
```

```python
angledegrees = linspace(0,90,num=200)
```

```python
rangeforangle100 = [rangeformula(100, x) for x in angledegrees]
rangeforangle200 = [rangeformula(200, x) for x in angledegrees]
plt.plot(angledegrees, rangeforangle100, label="100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label="200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.axis([0,100,0,4100])
plt.xlabel('Angle (º)')
plt.ylabel('Range (m)')
plt.show()
```

```python
rad = [radians(x) for x in angledegrees]
plt.plot(rad, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (º)')
plt.ylabel('Range (m)')
plt.axis([0,pi/2,0,1200])
plt.show()
```

```python
initialvelocity = range(0,201)
```

```python
rangefromvelocity = [rangeformula(x,45) for x in initialvelocity]
plt.plot(initialvelocity, rangefromvelocity)
plt.title('Range as a function of initial velocity at an angle of 45º')
plt.xlabel('Initial Velocty (m/s)')
plt.ylabel('Range (m)')
plt.axis([0,200,0,4100])
plt.show()
```

```python
rangefromvelocity15 = [rangeformula(x,15) for x in initialvelocity]
rangefromvelocity30 = [rangeformula(x,30) for x in initialvelocity]
rangefromvelocity45 = [rangeformula(x,45) for x in initialvelocity]
rangefromvelocity60 = [rangeformula(x,60) for x in initialvelocity]
rangefromvelocity75 = [rangeformula(x,75) for x in initialvelocity]
plt.plot(initialvelocity, rangefromvelocity15, label="15º")
plt.legend()
plt.plot(initialvelocity, rangefromvelocity30, label="30º")
plt.legend()
plt.plot(initialvelocity, rangefromvelocity45, label="45º")
plt.legend()
plt.plot(initialvelocity, rangefromvelocity60, label="60º")
plt.legend()
plt.plot(initialvelocity, rangefromvelocity75, label="75º")
plt.legend()
plt.title('Range as a function of initial velocity at various angles')
plt.axis([0,100,0,1200])
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (m)')
plt.show()
```
